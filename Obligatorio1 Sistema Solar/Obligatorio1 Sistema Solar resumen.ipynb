## Características de los planetas

```python
import numpy as np

import matplotlib.pyplot as plt

# Distancias promedio de los planetas al Sol en UA
distancias_ua = {
    "Mercurio": 0.39,
    "Venus": 0.72,
    "Tierra": 1.00,
    "Marte": 1.52,
    "Júpiter": 5.20,
    "Saturno": 9.58,
    "Urano": 19.22,
    "Neptuno": 30.05
}

# Diámetros promedio de los planetas en kilómetros
diametros_km = {
    "Mercurio": 4_879,
    "Venus": 12_104,
    "Tierra": 12_756,
    "Marte": 6_792,
    "Júpiter": 142_984,
    "Saturno": 120_536,
    "Urano": 51_118,
    "Neptuno": 49_528,
}

# Conversión de UA a kilómetros (1 UA = 149,597,870.7 km)
UA_A_KM = 149_597_870.7

# Mostrar las distancias en UA, kilómetros y radios en UA
print("Distancias de los planetas al Sol y sus radios:")
print(f"{'Planeta':<10} {'Distancia (UA)':<15} {'Distancia (km)':<20} {'Radio (UA)':<15}")
print("-" * 65)

for planeta, distancia_ua in distancias_ua.items():
    distancia_km = distancia_ua * UA_A_KM
    radio_km = diametros_km[planeta] / 2  # Calcular el radio a partir del diámetro
    radio_ua = radio_km / UA_A_KM  # Convertir el radio de kilómetros a UA
    print(f"{planeta:<10} {distancia_ua:<15} {distancia_km:<20,.2f} {radio_ua:<15.10f}")

# Calcular y mostrar el radio solar en UA
radio_solar_km = 695_700  # Radio del Sol en kilómetros
radio_solar_ua = radio_solar_km / UA_A_KM
print("\nRadio solar:")
print(f"El radio del Sol es {radio_solar_ua:.10f} UA")
```

## Conservación de la energía

```python
import numpy as np

import matplotlib.pyplot as plt

# Leer el archivo energias.txt
filename = "energias.txt"
data = np.loadtxt(filename)

# Extraer columnas de datos
T = data[:, 0]  # Energía cinética
V = data[:, 1]  # Energía potencial
E = data[:, 2]  # Energía mecánica total

# Crear un array de tiempo (en días)
days = np.arange(1, len(T) + 1)

# Representar las energías en función del tiempo
plt.figure(figsize=(7, 5))
plt.plot(days, T, label="Energía Cinética (T)", color="blue")
plt.plot(days, V, label="Energía Potencial (V)", color="green")
plt.plot(days, E, label="Energía Mecánica Total (E)", color="red")

# Configurar la gráfica
plt.title("Energías en función del tiempo")
plt.xlabel("Días")
plt.ylabel("Energía")
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

## Conservación del momento angular

```python
import matplotlib.pyplot as plt

def graficar_momento_angular(fichero):
    # Leer los valores del momento angular desde el fichero
    with open(fichero, 'r') as archivo:
        momento_angular = [float(line.strip()) for line in archivo]

    # Crear el eje de tiempo (un año con pasos de un día)
    tiempo = [i for i in range(len(momento_angular))]  # Tiempo en días

    # Calcular los límites del eje y basados en el rango de los datos
    valor_maximo = max(momento_angular)
    valor_minimo = min(momento_angular)
    margen = (valor_maximo - valor_minimo) * 0.1  # Margen del 10% del rango

    # Ajustar los límites del eje y
    limite_superior = valor_maximo + margen
    limite_inferior = valor_minimo - margen

    # Graficar el momento angular en función del tiempo
    plt.figure(figsize=(7, 5))
    plt.plot(tiempo, momento_angular, label="Momento Angular Total", color="blue")
    plt.xlabel("Tiempo (días)")
    plt.ylabel("Momento Angular Total (kg·m²/s)")
    plt.title("Momento Angular Total en función del Tiempo")
    plt.ylim(limite_inferior, limite_superior)  # Ajustar los límites del eje y
    plt.grid(True)
    plt.legend()
    plt.tight_layout()

    # Mostrar la gráfica
    plt.show()

# Llamar a la función con el fichero de datos
graficar_momento_angular("momento_angular_total.txt")

#ANOTACIÓN. Los valores del eje y hay que multiplicarlos por 10^43. Se ve entonces que las variaciones
#del momento angular con el tiempo son despreciables. Por tanto, se conserva. 
```
