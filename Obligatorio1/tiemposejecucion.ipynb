```python
import matplotlib.pyplot as plt
import numpy as np

# Nombres de los ficheros para cada opcin de compilacin
files = {
    "O1": "O1.txt",
    "O2": "O2.txt",
    "O3": "O3.txt",
    "OpenMP": "OpenMP.txt",
    "OpenMP+O2": "OpenMPO2.txt"
}

# Diccionario para almacenar los datos
data = {}

# Leer los datos de cada fichero
for label, filename in files.items():
    # Cada fichero debe tener dos columnas: tiempo de ejecucin y nmero de planetas
    datos = np.loadtxt(filename)
    tiempos = datos[:, 0]  # Primera columna: tiempos de ejecucin
    planetas = datos[:, 1]  # Segunda columna: nmero de planetas
    data[label] = (planetas, tiempos)

# Graficar los resultados
plt.figure(figsize=(10, 6))
for label, (planetas, tiempos) in data.items():
    plt.plot(planetas, tiempos, marker='o', label=label)

# Configurar la grfica
plt.title("Tiempo de ejecucin en funcin del nmero de planetas ( compilador AMD Clang)", fontsize=16)
plt.xlabel("Nmero de planetas (N)")
plt.ylabel("Tiempo de ejecucin (s)")
plt.legend()
plt.tight_layout()

# Mostrar la grfica
plt.show()
```
