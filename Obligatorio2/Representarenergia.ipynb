```python
import matplotlib.pyplot as plt

# Leer datos del archivo energias.txt
iteraciones = []
energias = []

with open("energias.txt", "r") as f:
    for linea in f:
        partes = linea.strip().split()
        if len(partes) == 2:
            iteraciones.append(int(partes[0]))
            energias.append(float(partes[1]))

# Graficar energía vs iteraciones uniendo los puntos
plt.figure(figsize=(8,5))
plt.plot(iteraciones, energias, label="Energía", color="blue", marker='o', markersize=3)
plt.xlabel("Pasos monte Carlo")
plt.ylabel("Energía")
plt.title("N=64, T=2.0, 10000 pasos monte Carlo.")
plt.legend()
plt.show()
```
